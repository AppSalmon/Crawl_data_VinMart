### Task 1: Add name and link of product

```python
df = pd.read_excel('Data_VinMart.xlsx')
```

```python
def get_link_name_product_one_page(link_page):
    driver.get(link_page)
    sleep(3)
    links = driver.find_elements('xpath', '/html/body/main/div/div/div/div/div/div/a')
    names = driver.find_elements('xpath', '/html/body/main/div/div/div/div/div/div/a/div/span')
    name_page = []
    link_page = []
    for name in names:
        name_page.append(name.text)
        print(name.text)
    for link in links:
        link_page.append(link.get_attribute('href'))
        print(link.get_attribute('href'))
    return name_page, link_page
```

```python
all_name = []
all_link = []
driver = webdriver.Chrome()
for i in tq.tqdm(range(1, 11)):
    link_page = f'https://vinmart.co/do-uong-100149/?trang={i}'
    print('Page:', i)
    name_one_page, link_one_page = get_link_name_product_one_page(link_page)
    sleep(2)
    all_name += name_one_page
    all_link += link_one_page
    print('len name:', len(all_name), 'len link:', len(all_link))
driver.quit()
```
